Preparação para o modelo de ML

```python
# A princípio para montar o modelo  que vai fazer a previsão  dos valores de aluguéis, precisamos realizar as etapas de limpeza e tratamento dos dados.
# Do df_rent vou retirar algumas features que contém dados que não ajudarão na boa performance do modelo, como as seguintes features: New,Negotiation Type e Property Type	.
# Estas features devem ser dropadas pois só tem dados com um tipo de valor ou quando não são  categóricos com um valor apenas.
# o objetivo é evitar underfitting e overfitting e confundir menos o modelo.
df_cleaned = df_rent.drop(["New","Property Type","Negotiation Type"],axis=1)
df_cleaned
```

```python
# Uma parte importante para nosso conjuno de dados é lidar com dados categóricos. Vamos trabalhar com a estratégia One hot in code. 
one_hot = pd.get_dummies(df_cleaned["District"])
one_hot
```

```python
# Vamos agora unir o df_cleaned com one_hot, pois precisamos realizar o preprocessamento de dados para assim então entrar no modelo de ML. Para isso temos que remover a coluna District e colocar o dataframe one hot, veja:
df = df_cleaned.drop('District', axis=1)
df = df.join(one_hot)
df 
```

```python
# Separando os dados em treino e teste:
from sklearn.model_selection import train_test_split
```

```python
# Definindo os dados de teste como a coluna Price e os dados de treino como todas as demais colunas exceto a Price, veja:
Y = df["Price"]
X = df.loc[:, df.columns != "Price" ]
```

```python
# chamando minha variável de aprendizado X:
X
```

```python
# chamando minha variável no  qual vou testar, cahmada Y:
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = 1)
```

```python
x_train
```

```python
y_train
```

```python
# Os modelos de previsão que utilizaremos serão dois(2), Regressão Linear e Random Forest Regressor. Primeiramente, iremos importar as bibliotecas abaixo.
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import numpy as np
```

```python
# Definiremos para as situações de treino e teste, duas métricas chamadas R² e RSME, que servirão para mostrar e descrever a qualidade de regressão gerada da série.
# Argumentos para Treino:
# Treino IA
lin_reg = LinearRegression()
lin_reg.fit (x_train, y_train)

rf_reg = RandomForestRegressor()
rf_reg.fit (x_train, y_train)
```

```python
# Argumentos para Teste:
# Teste IA
test_pred_lin = lin_reg.predict (x_test)
test_pred_rf = rf_reg.predict (x_test)

r2_lin = metrics.r2_score (y_test, test_pred_lin)
rmse_lin = np.sqrt(metrics.mean_squared_error(y_test, test_pred_lin))

print(f"R² da Regressão Linear: {r2_lin}")
print(f"RSME da Regressão Linear: {rmse_lin}")

r2_rf=metrics.r2_score(y_test, test_pred_rf)
rmse_rf=np.sqrt(metrics.mean_squared_error(y_test, test_pred_rf))

print(f"R² da Random Forest: {r2_rf}")
print(f"RSME do Random Forest: {rmse_rf}")
```

Análise gráfica e comparação.

Analisando os resultados, observamos que o valor de R² (valor que varia entre 0 e 1) da Random Forest é maior que o da Regressão Linear . Sendo assim, chegamos ao valor de 0.75 como número final, concluindo que o modelo de Random Forest é o mais assertivo para a predição de dados.

Em complemento às análises, iremos gerar em forma gráfica todos os pontos realizados, trazendo um comparativo entre os valores reais e preditos. Para isso, iremos executar os códigos da linha abaixo.

```python
df2_resultado = pd.DataFrame()
df2_resultado['y_teste'] = y_test
df2_resultado['y_previsao_rf'] = test_pred_rf
df2_resultado['y_previsao_lin'] = test_pred_lin
df2_resultado = df2_resultado.reset_index(drop=True)
fig = plt.figure(figsize=(15, 5))
sns.lineplot(data=df2_resultado)
plt.show()
display(df2_resultado)
```
